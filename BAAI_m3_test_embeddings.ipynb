```python
from sentence_transformers import SentenceTransformer, util

# Load the pre-trained model
model = SentenceTransformer('BAAI/bge-m3')

# Define the text to embed
text = "This is a sentence to be embedded."

# Encode the text into an embedding
embedding = model.encode(text)

# Print the embedding
print(embedding)
```

```python
#!pip install beautifulsoup4==4.12.2 requests==2.31.0
from bs4 import BeautifulSoup
import requests

# Define the URL of the webpage
url = "https://www.vivoscuola.it/content/view/full/233786"

# Fetch the webpage content
response = requests.get(url)
html_content = response.text

# Create a BeautifulSoup object
soup = BeautifulSoup(html_content, "html.parser")

# Find elements with classes starting with "Accordion-header" and "Accordion-panel"
header_elements = soup.find_all(lambda tag: tag.name == 'div' and tag.get('class') and tag.get('class')[0].startswith('Accordion-header'))
panel_elements = soup.find_all(lambda tag: tag.name == 'div' and tag.get('class') and tag.get('class')[0].startswith('Accordion-panel'))

# Extract and print the inner text
for element in header_elements:
    print(element.get_text())

for element in panel_elements:
    print(element.get_text())
```

```python
import chromadb
from chromadb.utils import embedding_functions

# Initialize ChromaDB
client = chromadb.Client()

# Create an in-memory collection
collection = client.create_collection(name="my_collection")

# Assuming you have header_elements and panel_elements from the previous code
documents = [element.get_text() for element in header_elements + panel_elements]
embeddings = model.encode(documents).tolist()

# Add data to the collection
collection.add(
    documents=documents,
    embeddings=embeddings,
    metadatas=[{"source": "header"} for _ in header_elements] + [{"source": "panel"} for _ in panel_elements],
    ids=[f"doc_{i}" for i in range(len(documents))],  # Optional: Assign custom IDs
)
```

```python
collection.count()
```

```python
query_text = "quali sono i requisiti per partecipare al concorso?"
query_embedding = model.encode(query_text).tolist()

results = collection.query(
    query_embeddings=query_embedding,
    n_results=4,
)
print(results)
print("*******************")
# Print the results
for result in results["documents"][0]:
    print(result)
    print("================")
```
